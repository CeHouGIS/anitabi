```python
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
```

```python
point_dir = glob("/data_nas/cehou/anitabi/metadata/*/detailed_points_metadata.csv")
anime_metadata = glob("/data_nas/cehou/anitabi/metadata/*/anime_metadata.csv")
```

```python
for i, df_name in tqdm(enumerate(anime_metadata)):
    if i == 0:
        df_all = pd.read_csv(df_name)
        df_all["anime_id"] = df_name.split("/")[-2]
    else:
        df = pd.read_csv(df_name)
        df["anime_id"] = df_name.split("/")[-2]
        df_all = pd.concat([df_all, df], axis=0)

df_all.reset_index(drop=True, inplace=True)    
# df_all.to_pickle("/data_nas/cehou/anitabi/data/anime_metadata.pkl")
```
